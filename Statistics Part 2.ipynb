1. What is hypothesis testing in statistics?
- Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on a sample of data. It involves testing an assumption (called a hypothesis) about a population parameter.

#Key Concepts:
#Null Hypothesis (H₀):
This is the default assumption or claim. It usually states that there is no effect or no difference.
Example: "The average height of students is 170 cm."

#Alternative Hypothesis (H₁ or Ha):
This is what you want to test for—an opposing claim to the null.
Example: "The average height of students is not 170 cm."

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis and the alternative hypothesis are two opposing statements used in hypothesis testing to assess whether there's enough evidence in a sample of data to infer something about a population.

#Null Hypothesis (H₀)
It is the default assumption or status quo.

It suggests that there is no effect, no difference, or no change.

It's the hypothesis that you assume true until evidence suggests otherwise.

#Examples:

A new drug has no effect: H₀: μ = μ₀

A coin is fair: H₀: p = 0.5

There is no difference in test scores between two groups: H₀: μ₁ = μ₂

# Alternative Hypothesis (H₁ or Ha)
This is the hypothesis you want to test or prove.

It represents a new effect, difference, or change from the status quo.

If the data provides strong enough evidence against H₀, you accept H₁.

#Examples:

The drug is effective: H₁: μ ≠ μ₀ or μ > μ₀ or μ < μ₀

The coin is biased: H₁: p ≠ 0.5

Test scores are different between groups: H₁: μ₁ ≠ μ₂

# Key Differences
Feature - Null Hypothesis (H₀) - Alternative Hypothesis (H₁)
Assumption - No change or effect - There is a change or effect
Purpose - Baseline to test against  -  Represents the claim you are testing
Decision    You reject or fail to reject H₀ You support H₁ only if H₀ is rejected
Default Assumed true until evidence shows otherwise Considered if data contradicts H₀

#Example Scenario:
You're testing if a new teaching method improves student performance.

H₀: The new method has no effect (mean score = 75) → μ = 75

H₁: The new method improves scores → μ > 75

If test results show significantly higher scores (with a low p-value), you reject H₀ and accept H₁.

3. What is the significance level in hypothesis testing, and why is it important?
- The significance level in hypothesis testing, often denoted by α (alpha), is the threshold for deciding whether to reject the null hypothesis (H₀). It represents the probability of making a Type I error, which is rejecting H₀ when it is actually true.
#Key Points About Significance Level (α):
It’s chosen by the researcher before conducting the test.

Common values are:

0.05 (5% chance of Type I error)

0.01 (1%)

0.10 (10%)

#Why Is It Important?
Controls the risk of false positives (Type I error):
By setting α = 0.05, you’re accepting a 5% risk of incorrectly rejecting a true null hypothesis.

Sets the decision boundary:
If your p-value ≤ α, you reject H₀.
If your p-value > α, you fail to reject H₀.

Balances caution and sensitivity:
A lower α (like 0.01) makes it harder to reject H₀ (more cautious). A higher α (like 0.10) makes it easier but increases the risk of a false positive.

# Example:
You're testing a new medication.

H₀: The medication has no effect.

H₁: The medication is effective.

You choose α = 0.05.

Your test gives a p-value of 0.03.

Since 0.03 < 0.05, you reject H₀ and conclude the medication may be effective.

But: there’s a 5% chance this result happened by random chance—this is the risk you accepted by setting α = 0.05.

4. What does a P-value represent in hypothesis testing?
- The p-value in hypothesis testing represents the probability of observing your sample data—or something more extreme—if the null hypothesis (H₀) is true.

 #In Simple Terms:
A p-value tells you how surprising your results are under the assumption that H₀ is true.

#Small p-value (typically ≤ 0.05):
Strong evidence against the null hypothesis → Reject H₀

#Large p-value (> 0.05):
Weak evidence against H₀ → Fail to reject H₀

#Example:
You're testing whether a coin is fair:

H₀: The coin is fair (p = 0.5)

You flip it 100 times and get 70 heads.

You compute the p-value, which turns out to be 0.002.

Interpretation:
If the coin really were fair, there's only a 0.2% chance of getting 70 or more heads just by luck.
That’s very unlikely, so you reject H₀ — the coin probably isn't fair.

5- How do you interpret the P-value in hypothesis testing?
- How to Interpret the P-value in Hypothesis Testing
The p-value helps you decide whether your sample data provides enough evidence to reject the null hypothesis (H₀).

# Core Interpretation:
The p-value is the probability of obtaining a result at least as extreme as the one observed, assuming H₀ is true.

#Example Interpretation:
Scenario:
A researcher tests whether a new drug lowers blood pressure more than the current standard.

H₀: The new drug is no more effective than the standard.

H₁: The new drug is more effective.

Significance level: α = 0.05

Test result: p = 0.02

#Interpretation:
Since 0.02 < 0.05, the result is statistically significant.

You reject H₀, concluding there's evidence the new drug is more effective.

There is only a 2% chance of getting this result if H₀ were true.

# Important Notes:
A small p-value means the data is unlikely under H₀ → suggests H₀ may not be true.

A large p-value means the data is plausible under H₀ → no reason to reject H₀.

P-value ≠ probability that H₀ is true or false.

6. What are Type 1 and Type 2 errors in hypothesis testing?
- In hypothesis testing, Type I and Type II errors represent two different kinds of mistakes you can make when making a decision based on sample data.

# Type I Error (False Positive)
Definition: Rejecting the null hypothesis (H₀) when it is actually true.

You think there is an effect when there isn’t one.

#Symbol:

α (alpha) — this is your significance level (e.g., 0.05).

#Example:

You conclude a new drug works better than the old one, but in reality, there's no difference.

You're wrongly rejecting a true null hypothesis.

#Type II Error (False Negative)
Definition: Failing to reject H₀ when the alternative hypothesis (H₁) is actually true.

You miss a real effect.

#Symbol: β (beta)

#Power of the test:
1 − β → The probability of correctly rejecting H₀ when H₁ is true.

#Example:
 You conclude the new drug is no better than the old one, but it actually is.

 You're wrongly keeping a false null hypothesis.


#Visual Analogy (Courtroom Example):
Hypothesis Testing  Legal System Equivalent
H₀ (null)   Defendant is innocent
H₁ (alternative)    Defendant is guilty
Type I Error    Convict an innocent person
Type II Error   Let a guilty person go free

#How to Manage the Errors:
Lowering α reduces Type I errors but increases risk of Type II errors.

Increasing sample size helps reduce both errors.

Balancing α and β depends on context (e.g., in medicine, avoiding Type I errors is often more important).

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- Difference Between One-Tailed and Two-Tailed Tests in Hypothesis Testing
The main difference lies in the direction of the test and how the alternative hypothesis (H₁) is stated.
#One-Tailed Test
Tests for a directional effect — either greater than or less than.

The critical region (where you reject H₀) is only on one side of the distribution.
#Example:
You want to test if a new teaching method increases test scores.

H₀: μ = 75

H₁: μ > 75 (right-tailed test)

You’re only looking for evidence that the average is greater, not different in both directions.

# Used When:
You have a specific direction in mind (e.g., "faster," "higher," "less than").

#Two-Tailed Test
Tests for a non-directional difference — whether the value is different (either higher or lower).

The critical region is split between both tails of the distribution.

#Example:
You want to test if a new process changes the average delivery time (could be faster or slower).

H₀: μ = 5 days

H₁: μ ≠ 5 days

You're looking for any significant change, in either direction.

# Used When:
You're testing for any difference, not just an increase or decrease.

11. What is the difference between a Z-test and a T-test?

- Difference Between a Z-Test and a T-Test
Both Z-tests and T-tests are statistical methods used to compare sample data against a population or another sample — but they differ based on sample size, variance knowledge, and distribution used.

#Example Scenarios:
#Use a Z-Test When:
You're testing whether the average battery life of 100 phones is different from 10 hours.

You know the population standard deviation (e.g., σ = 1.2 hours).

Sample size is large (n = 100).

#Use a T-Test When:
You have a sample of 12 patients and want to test if a drug lowers blood pressure compared to a known average.

The population standard deviation is unknown.

Sample size is small (n = 12).
#Why It Matters:
Using the wrong test can lead to incorrect conclusions:

Using a Z-test with a small sample can underestimate variability, increasing the risk of Type I error.

Using a T-test with a large sample is okay — it becomes similar to a Z-test as sample size increases.

The Z-test and T-test are both statistical tools used to test hypotheses about population means, but they are used under different conditions. Despite these differences, they are closely related and serve similar purposes.

"Do these groups come from populations with the same mean?"

#Assumptions of ANOVA:
Assumption	Description
1. Independence	Observations must be independent (e.g., different subjects in each group).
2. Normality	The data in each group should be approximately normally distributed.
3. Homogeneity of variances	All groups should have equal variances (can be tested with Levene’s test or Bartlett’s test).

Note: ANOVA is fairly robust to violations of normality if group sizes are large and equal.

- There are three main types of ANOVA tests, each designed for specific experimental designs and research questions. Here's a breakdown:

Types of effects tested:

Main effect of Factor A (e.g., teaching method)

Main effect of Factor B (e.g., gender)

Purpose:
Used when the same subjects are measured multiple times (i.e., dependent or correlated samples). Often used in time-series or longitudinal studies.

# Example:
Measure blood pressure of patients before, during, and after treatment.

Assumptions:

Sphericity (variances of the differences are equal)

Normality of differences

Data points from the same subject across conditions

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
def bernoulli_pdf(x, p):
    """
    Calculate the probability mass function (PDF) of a Bernoulli distribution.

    Parameters:
    x (int): The outcome (0 or 1).
    p (float): The probability of success (1), where 0 <= p <= 1.

    Returns:
    float: The probability of observing x.
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1 inclusive")

    return p**x * (1 - p)**(1 - x)

# Example usage:
print(bernoulli_pdf(1, 0.7))  # Output: 0.7
print(bernoulli_pdf(0, 0.7))  # Output: 0.3
```

```python
#6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(x, a, b):
    """
    Calculate the CDF of a discrete uniform distribution.

    Parameters:
    x : array-like or float : values at which to evaluate the CDF
    a : int : lower bound (inclusive)
    b : int : upper bound (inclusive)

    Returns:
    array-like or float : CDF values
    """
    x = np.asarray(x)
    cdf = np.piecewise(
        x,
        [x < a, (x >= a) & (x <= b), x > b],
        [0, lambda x: (np.floor(x) - a + 1) / (b - a + 1), 1]
    )
    return cdf

# Parameters
a, b = 1, 6  # Like a fair die
x_values = np.linspace(a - 2, b + 2, 1000)
cdf_values = discrete_uniform_cdf(x_values, a, b)

# Plot CDF
plt.figure(figsize=(8, 5))
plt.plot(x_values, cdf_values, drawstyle='steps-post', color='blue', label='CDF')
plt.title(f'Cumulative Distribution Function\nDiscrete Uniform Distribution (a={a}, b={b})')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5)
plt.axhline(1, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them?
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores from a dataset and plot them.

    Parameters:
    data (array-like): Input numeric dataset.
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data)

    if std == 0:
        raise ValueError("Standard deviation is zero. Z-scores cannot be computed.")

    z_scores = (data - mean) / std

    # Plotting
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    # Histogram of Z-scores
    axs[0].hist(z_scores, bins=30, color='skyblue', edgecolor='black')
    axs[0].set_title('Histogram of Z-Scores')
    axs[0].set_xlabel('Z-score')
    axs[0].set_ylabel('Frequency')
    axs[0].grid(True, linestyle='--', alpha=0.6)

    # Scatter plot of original index vs Z-score
    axs[1].scatter(range(len(z_scores)), z_scores, color='purple', alpha=0.6)
    axs[1].axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    axs[1].set_title('Z-Scores Across Data Points')
    axs[1].set_xlabel('Index')
    axs[1].set_ylabel('Z-score')
    axs[1].legend()
    axs[1].grid(True, linestyle='--', alpha=0.6)

    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage with random normal data
if __name__ == "__main__":
    sample_data = np.random.normal(loc=50, scale=10, size=1000)
    z = calculate_and_plot_z_scores(sample_data)
```

```python
#10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
import numpy as np
import matplotlib.pyplot as plt

def demonstrate_clt(sample_size=30, num_samples=1000, lam=1.0):
    """
    Demonstrates the Central Limit Theorem using an exponential distribution.

    Parameters:
    sample_size (int): Size of each sample to draw.
    num_samples (int): Number of samples to draw.
    lam (float): Lambda parameter for exponential distribution (rate).
    """
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.exponential(scale=1/lam, size=sample_size)
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Plotting the results
    plt.figure(figsize=(10, 6))

    # Histogram of original population (for reference)
    population = np.random.exponential(scale=1/lam, size=10000)
    plt.subplot(1, 2, 1)
    plt.hist(population, bins=50, color='salmon', edgecolor='black', alpha=0.7)
    plt.title('Original Exponential Distribution')
    plt.xlabel('Value')
    plt.ylabel('Frequency')

    # Histogram of sample means
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, color='skyblue', edgecolor='black', density=True)
    plt.title(f'Distribution of Sample Means (n={sample_size})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

    plt.tight_layout()
    plt.show()

# Example usage
if __name__ == "__main__":
    demonstrate_clt(sample_size=50, num_samples=1000)
```

```python
#15.Implement hypothesis testing using Z-statistics for a sample dataset?
import numpy as np
from scipy.stats import norm

def z_test(sample, mu_0, sigma, alternative='two-sided', alpha=0.05):
    """
    Perform a Z-test for the population mean.

    Parameters:
    sample (array-like): Sample data
    mu_0 (float): Hypothesized population mean (null hypothesis)
    sigma (float): Population standard deviation (known)
    alternative (str): 'two-sided', 'greater', or 'less'
    alpha (float): Significance level

    Returns:
    Dictionary with test results
    """
    sample = np.array(sample)
    x_bar = np.mean(sample)
    n = len(sample)
    se = sigma / np.sqrt(n)
    z_stat = (x_bar - mu_0) / se

    # p-value calculation
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_stat)))
        z_crit = norm.ppf(1 - alpha/2)
        reject = abs(z_stat) > z_crit
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
        z_crit = norm.ppf(1 - alpha)
        reject = z_stat > z_crit
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
        z_crit = norm.ppf(alpha)
        reject = z_stat < z_crit
    else:
        raise ValueError("Invalid alternative hypothesis")

    return {
        'sample_mean': x_bar,
        'z_statistic': z_stat,
        'p_value': p_value,
        'reject_null': reject,
        'z_critical': z_crit
    }

# Example usage
if __name__ == "__main__":
    # Simulate a dataset
    np.random.seed(0)
    sample_data = np.random.normal(loc=52, scale=10, size=50)  # sample from population

    # Hypothesis: Population mean is 50
    result = z_test(sample_data, mu_0=50, sigma=10, alternative='two-sided', alpha=0.05)

    print("Z-Test Results:")
    for key, value in result.items():
        print(f"{key}: {value:.4f}" if isinstance(value, float) else f"{key}: {value}")
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result?
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a dataset.

    Parameters:
    - data: list or array of numeric values
    - confidence: confidence level (default: 0.95)

    Returns:
    - Tuple: (lower_bound, upper_bound, mean)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t-critical value

    margin = t_crit * std_err
    lower = mean - margin
    upper = mean + margin

    return lower, upper, mean

# Example usage
if __name__ == "__main__":
    # Simulated dataset
    np.random.seed(1)
    sample = np.random.normal(loc=100, scale=15, size=50)

    lower, upper, mean = confidence_interval(sample, confidence=0.95)

    print(f"Sample Mean: {mean:.2f}")
    print(f"95% Confidence Interval: [{lower:.2f}, {upper:.2f}]")
    print("\nInterpretation:")
    print("We are 95% confident that the true population mean lies within this interval.")
```

```python
#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
from scipy import stats

def generate_data_and_ci(mu=50, sigma=10, n=100, confidence=0.95):
    """
    Generate normal data, calculate and interpret confidence interval for the mean.

    Parameters:
    mu (float): True mean of the normal distribution
    sigma (float): True std dev of the normal distribution
    n (int): Sample size
    confidence (float): Confidence level

    Returns:
    tuple: (sample_data, ci_lower, ci_upper, sample_mean)
    """
    # Generate random data from normal distribution
    data = np.random.normal(loc=mu, scale=sigma, size=n)

    # Calculate sample mean and SEM
    sample_mean = np.mean(data)
    sem = stats.sem(data)

    # t-critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of error
    margin_error = t_crit * sem

    # Confidence interval
    ci_lower = sample_mean - margin_error
    ci_upper = sample_mean + margin_error

    return data, ci_lower, ci_upper, sample_mean

if __name__ == "__main__":
    # Parameters
    true_mu = 50
    true_sigma = 10
    sample_size = 100
    conf_level = 0.95

    data, lower, upper, mean = generate_data_and_ci(mu=true_mu, sigma=true_sigma, n=sample_size, confidence=conf_level)

    print(f"Generated {sample_size} data points from N({true_mu}, {true_sigma**2})")
    print(f"Sample Mean: {mean:.2f}")
    print(f"{conf_level*100:.0f}% Confidence Interval for the Mean: [{lower:.2f}, {upper:.2f}]")
    print("\nInterpretation:")
    print(f"We are {conf_level*100:.0f}% confident that the true population mean ({true_mu}) lies within this interval.")
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std):
    # Calculate sample statistics
    sample_mean = np.mean(sample)
    n = len(sample)

    # Compute Z-score
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Interpretation
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")

    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

    return z, p_value

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=102, scale=15, size=50)  # Sample with mean ≈ 102
population_mean = 100
population_std = 15

z_test(sample_data, population_mean, population_std)
```

```python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate standard error
    se = pop_std / np.sqrt(n)

    # Calculate z-score
    z = (sample_mean - pop_mean) / se

    # Calculate two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample mean: {sample_mean:.4f}")
    print(f"Population mean: {pop_mean}")
    print(f"Z-score: {z:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis: significant difference.")
    else:
        print("Fail to reject the null hypothesis: no significant difference.")

    return z, p_value

# Example usage:
np.random.seed(0)
sample_data = np.random.normal(loc=102, scale=15, size=50)
population_mean = 100
population_std = 15  # Known population std dev

one_sample_z_test(sample_data, population_mean, population_std)
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)
    z = (sample_mean - pop_mean) / se

    # Critical z-values for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)

    # p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample mean: {sample_mean:.4f}")
    print(f"Z-score: {z:.4f}")
    print(f"Critical z-value: ±{z_critical:.4f}")
    print(f"P-value: {p_value:.4f}")

    if abs(z) > z_critical:
        print("Reject the null hypothesis: significant difference.")
    else:
        print("Fail to reject the null hypothesis: no significant difference.")

    # Plotting
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10,6))
    plt.plot(x, y, label='Standard Normal Distribution')

    # Shade rejection regions
    plt.fill_between(x, 0, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
    plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.3)

    # Plot test statistic
    plt.axvline(z, color='blue', linestyle='--', label=f'Z-test Statistic = {z:.2f}')

    plt.title('Two-tailed Z-test Decision Regions')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z, p_value

# Example usage
np.random.seed(42)
sample = np.random.normal(loc=102, scale=15, size=50)
population_mean = 100
population_std = 15

two_tailed_z_test(sample, population_mean, population_std)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_type1_type2_errors(pop_mean, pop_std, sample_size, alpha, alt_mean):
    """
    Visualize Type I and Type II errors for a one-sample Z-test.

    Parameters:
    - pop_mean: Mean under the null hypothesis
    - pop_std: Known population standard deviation
    - sample_size: Sample size (n)
    - alpha: Significance level (Type I error rate)
    - alt_mean: Mean under the alternative hypothesis
    """
    se = pop_std / np.sqrt(sample_size)

    # Critical z-value for two-tailed test
    z_crit = norm.ppf(1 - alpha/2)

    # Critical values on x-axis (sample mean scale)
    crit_low = pop_mean - z_crit * se
    crit_high = pop_mean + z_crit * se

    # Define x range to plot both null and alternative distributions
    x = np.linspace(pop_mean - 4*se, alt_mean + 4*se, 1000)

    # Null hypothesis distribution (sampling distribution of sample mean under H0)
    null_dist = norm.pdf(x, loc=pop_mean, scale=se)

    # Alternative hypothesis distribution (sampling distribution under H1)
    alt_dist = norm.pdf(x, loc=alt_mean, scale=se)

    plt.figure(figsize=(10,6))

    # Plot null hypothesis distribution
    plt.plot(x, null_dist, label='Null Hypothesis (H0)', color='blue')
    # Plot alternative hypothesis distribution
    plt.plot(x, alt_dist, label='Alternative Hypothesis (H1)', color='green')

    # Shade rejection regions under H0 (Type I error regions)
    plt.fill_between(x, 0, null_dist, where=(x <= crit_low), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, 0, null_dist, where=(x >= crit_high), color='red', alpha=0.3)

    # Shade Type II error region under H1 (where we fail to reject H0, i.e. between critical values)
    plt.fill_between(x, 0, alt_dist, where=(x > crit_low) & (x < crit_high), color='orange', alpha=0.3, label='Type II Error (β)')

    # Plot critical values as vertical lines
    plt.axvline(crit_low, color='black', linestyle='--', label='Critical Values')
    plt.axvline(crit_high, color='black', linestyle='--')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Calculate beta (Type II error)
    beta_low = norm.cdf(crit_high, loc=alt_mean, scale=se) - norm.cdf(crit_low, loc=alt_mean, scale=se)
    beta = beta_low

    print(f"Significance level (Type I error, α): {alpha}")
    print(f"Type II error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage:
plot_type1_type2_errors(
    pop_mean=100,
    pop_std=15,
    sample_size=30,
    alpha=0.05,
    alt_mean=110  # True mean under alternative hypothesis
)
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_crit * std_err

    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample mean: {mean:.4f}")
    print(f"{int(confidence * 100)}% Confidence Interval: ({lower:.4f}, {upper:.4f})")

    return lower, upper

# Example usage:
np.random.seed(1)
sample_data = np.random.normal(loc=50, scale=10, size=30)
confidence_interval(sample_data)
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    n = len(data)
    sample_std = np.std(data, ddof=1)
    std_err = sample_std / np.sqrt(n)

    # Get critical t-value
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    # Calculate margin of error
    margin_error = t_crit * std_err

    print(f"Sample Size: {n}")
    print(f"Sample Mean: {np.mean(data):.4f}")
    print(f"Sample Standard Deviation: {sample_std:.4f}")
    print(f"Standard Error: {std_err:.4f}")
    print(f"Margin of Error at {int(confidence * 100)}% confidence: {margin_error:.4f}")

    return margin_error

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=25)
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior, sensitivity, false_positive_rate):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Parameters:
    - prior: P(Disease) - prior probability of having the disease
    - sensitivity: P(Pos | Disease) - true positive rate
    - false_positive_rate: P(Pos | No Disease)

    Returns:
    - posterior: P(Disease | Pos)
    """
    p_no_disease = 1 - prior
    # Total probability of a positive test (evidence)
    evidence = sensitivity * prior + false_positive_rate * p_no_disease

    # Apply Bayes' Theorem
    posterior = (sensitivity * prior) / evidence

    print(f"Posterior probability of disease given a positive test: {posterior:.4f}")
    return posterior

# Example usage
bayes_theorem(prior=0.01, sensitivity=0.99, false_positive_rate=0.05)
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test.

    Parameters:
    - observed: 2D array-like contingency table (rows: categories of variable 1, columns: variable 2)

    Returns:
    - expected: DataFrame of expected frequencies
    """
    observed = np.array(observed)
    row_totals = observed.sum(axis=1, keepdims=True)
    col_totals = observed.sum(axis=0, keepdims=True)
    total = observed.sum()

    expected = row_totals @ col_totals / total  # Matrix multiplication

    return pd.DataFrame(expected,
                        index=[f'Row{i+1}' for i in range(observed.shape[0])],
                        columns=[f'Col{j+1}' for j in range(observed.shape[1])])

# Example observed data
observed_data = [
    [25, 30, 20],
    [15, 25, 35]
]

expected_df = calculate_expected_frequencies(observed_data)
print("Expected Frequencies:\n", expected_df)
```
